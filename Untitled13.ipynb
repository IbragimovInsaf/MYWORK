```python
########## Подготовим датасет для обучения
top_five = ['Bundesliga', 'Ligue 1', 'La Liga', 'Premier League', 'Serie A']
df_male_players["is_top_five"] = df_male_players['league_name'].apply(lambda x: 1 if x in top_five else 0)

### Проверка разметки
df_male_players[['league_name', 'is_top_five']].drop_duplicates()
```

```python
### Предобработка необходимых столбцов
import re
set = ['ls', 'st', 'rs', 'lw',
       'lf', 'cf', 'rf', 'rw', 'lam', 'cam', 'ram', 'lm', 'lcm', 'cm', 'rcm',
       'rm', 'lwb', 'ldm', 'cdm', 'rdm', 'rwb', 'lb', 'lcb', 'cb', 'rcb', 'rb',
       'gk']
def changenumber(string):
  if isinstance(string, str):
    if '-' in string:
      s = re.split(r'\-', string)
      return int(s[0]) - int(s[1])
    else:
      return sum([int(elem) for elem in re.split(r'\+|\*', string)])
  else:
    return int(string)

df_male_players.fillna(0, inplace=True)
for elem in set:
  df_male_players[elem] = df_male_players[elem].apply(lambda x :changenumber(x))
```

```python
### Разделим датасет по годам
df_male_players_15 = pd.DataFrame(df_male_players[df_male_players['fifa_version'] == 15]) 
df_male_players_16 = pd.DataFrame(df_male_players[df_male_players['fifa_version'] == 16]) 
df_male_players_17 = pd.DataFrame(df_male_players[df_male_players['fifa_version'] == 17]) 
df_male_players_18 = pd.DataFrame(df_male_players[df_male_players['fifa_version'] == 18])
df_male_players_19 = pd.DataFrame(df_male_players[df_male_players['fifa_version'] == 19]) 
df_male_players_20 = pd.DataFrame(df_male_players[df_male_players['fifa_version'] == 20]) 
df_male_players_21 = pd.DataFrame(df_male_players[df_male_players['fifa_version'] == 21]) 
df_male_players_22 = pd.DataFrame(df_male_players[df_male_players['fifa_version'] == 22])
df_male_players_23 = pd.DataFrame(df_male_players[df_male_players['fifa_version'] == 23])
```

```python
import sqlite3
conn = sqlite3.connect("test.sqlite")
cur = conn.cursor()
cur.execute("""
CREATE TABLE players_23_ (
    player_id integer,
    is_top_five_f integer
)
""")
cur.executemany("INSERT INTO players_23_ VALUES (?, ?)", zip(df_male_players_23['player_id'],df_male_players_23['is_top_five']))
conn.commit()
df = pd.read_sql(
    "SELECT * FROM players_23_",
    conn,
)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, DecisionTreeRegressor
from sklearn.metrics import mean_squared_error
from scipy.optimize import minimize_scalar
```

```python
players_15_x_23 = df_male_players_15.merge(df, on='player_id', how='inner')
todrop.append('is_top_five')
todrop.append('is_top_five_f')
```

```python
# X = players_15_x_23[players_15_x_23.drop(columns = todrop,axis = 1).columns]
X = players_15_x_23[players_15_x_23.drop(columns = todrop,axis = 1).columns]
y = players_15_x_23['is_top_five_f']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=115)
```

```python
from sklearn.metrics import roc_curve, auc
from sklearn. metrics import precision_recall_curve
from sklearn.metrics import accuracy_score
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(max_features=1, n_estimators=10, random_state=190654543)

rf.fit(X_train, y_train)

#accuracy на тесте
print(accuracy_score(y_test,rf.predict(X_test)))
#accuracy на трене (показатель переобучаемости)
print(accuracy_score(y_train,rf.predict(X_train)))
```

```python
from sklearn.ensemble import BaggingClassifier
from sklearn.tree import DecisionTreeClassifier

bc = BaggingClassifier().fit(X_train, y_train)

#accuracy на тесте
print(accuracy_score(y_test,bc.predict(X_test)))
#accuracy на трене (показатель переобучаемости)
print(accuracy_score(y_train,bc.predict(X_train)))
```

```python
from sklearn.ensemble import GradientBoostingClassifier
bc2 = BaggingClassifier(GradientBoostingClassifier(n_estimators=100)).fit(X_train, y_train)

#accuracy на тесте
print(accuracy_score(y_test,bc2.predict(X_test)))
#accuracy на трене (показатель переобучаемости)
print(accuracy_score(y_train,bc2.predict(X_train)))
```
